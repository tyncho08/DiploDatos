### Pasos Fronterizos de Argentina

Cargamos las librerías necesarias, en este caso la librería __pandas__ y __gmplot__

```python
%matplotlib inline
import pandas as pd
from gmplot import gmplot
from IPython.core.display import display, HTML
```

Leemos los datos de los pasos fronterizos, para ello debemos leer el archivo en formato __json__ llamado __pasos.json__

```python
pasos = pd.read_json("../input/pasos.json")
ubicacion = pasos.location
pasos
```

Ahora debemos hacer uso de la librería __gmplot__ para poder marcar los puntos extraídos del archivo anterior

```python
# Mapa
gmap = gmplot.GoogleMapPlotter(-34, -64, 4.5)

# Puntos tipo Scatter
top_attraction_lats, top_attraction_lons = zip(*ubicacion)
gmap.scatter(top_attraction_lats, top_attraction_lons, 'cornflowerblue', size=20, marker=True)

# Grafico en el mapa
gmap.draw("pasos_fronterizos_de_Argentina.html")
```

El mapa resultante se encuentra en el archivo __pasos_fronterizos_de_Argentina.html__
