### Analizar texto

Importamos las librerías que se utilizarán, creamos una función para leer el texto y cargamos el modelo de spacy que utilizaremos para procesar el texto:

```python
# <!-- collapse=True -->
# Importando las librerías que vamos a utilizar
import matplotlib.pyplot as plt 
import textacy
import spacy
from textacy.datasets import Wikipedia
from collections import Counter, defaultdict
import warnings; warnings.simplefilter('ignore')

# graficos incrustados
%matplotlib inline

# función auxiliar
def leer_texto(texto):
    """Funcion auxiliar para leer un archivo de texto"""
    with open(texto, 'r') as text:
        return text.read()
    
# Cargando el modelo en español de spacy
nlp = spacy.load('es_core_news_md')
```

Ahora cargamos el texto que nos interesa y lo procesamos con el modelo elegido. Cabe aclarar que el encabezado y la cola del tecto puestas por the gutenberg project han sido extraídos manualmente. En el caso que se quisiera aplicar esto a muchos textos hay que encontrar el patrón que se repite en el encabezado y la cola de los textos y eliminarlos antes de que los textos sean procesados.

```python
texto = leer_texto('../input/LaIslaDelTesoroFilt.txt')
texto_procesado = nlp(texto)
```

Ahora procedemos a contar las palabras excluyendo las stop words y los signos de puntuación. El filtrado por signos de puntuación no alcanza para eliminar todos los símbolos (quedan saltos de líneas, espacios en blanco y algunos símbolos) con lo cual además nos fijamos si nuestro string posee todo símbolos alfabéticos y nos quedamos solo con esas palabras. Además notamos que había palabras que si bien no estaban en stop words no nos aportaban nada al análisis (y, Y, á, ó), debido a esto luego de analizar los resultados se decidió aplicar un filtro dejándo pasar solo las palabras de 3 letras o más. Finalmente ordenamos el resultado en base a la frecuencia de las mismas.

```python
words = [token.text for token in texto_procesado if token.is_stop != True and token.is_punct != True and token.is_alpha == True and len(token.text) > 2]
word_freq = Counter(words)
common_words = word_freq.most_common()
```

La lista de palabras ordenadas por frecuencia se presenta a continuación:

```python
common_words
```

A continuación se encontrarán los personajes contenidos en la historia de nuestro texto. Gracias al análisis que hemos hecho a través de la librería spacy la función texto_procesado.ents nos devolverá las diferentes entidades que encuentra en el texto. Estas entidades pueden ser conjuntos de palabras y por lo tanto son importantes para poder identificar los nombres completos (en caso de haber un nombre y apellido). Accediendo al label de cada entidad contenida en nuestro texto podremos saber si se trata o no de un personaje. Pero nos encontramos con falsos positivos en los resultados y debimos aplicar ciertos procedimientos para filtrarlos. Observando los resultados encontramos diversa cantidad de símbolos a los costados de nuestras entidades, así como el string '-PRON-' el cual no aportaba nada, con lo cual se decidió filtrar todos estos símbolos de nuestros strings. Además se quitaron todos los saltos de línea de las entidades y se eliminaron las que estaban vacías. Nuevamente descartamos todas las entidades que tuvieran 2 caracteres o menos, ya que nos ayudaba a filtrar muchas palabras basura y los nombres de personajes en general cuentan con mas caracteres. Por último notamos que teníamos muchos verbos que eran considerados personajes, para solucionar esto a nuestro resultado lo volvimos a pasar por nuestro modelo de spacy y recorrimos cada personaje revisando si contenía verbos, en caso de tenerlos lo descartabamos. Con todo esto pudimos filtrar la mayoría de las entidades basura infiltradas en nuestros personajes, no obstante quedaron bastantes.

```python
personajes = []
for ent in texto_procesado.ents:
    if ent.label_ == 'PER' and ent.lemma_ != '' and ent.lemma_.replace('-PRON-','').strip() != '' and len(ent.lemma_.replace('\n','').strip().strip('.').strip('_').strip('¡').strip('!').strip('¿').strip('?'))>2:
        substring = texto[ent.start_char: ent.end_char].replace('\n','').strip().strip('.').strip('_').strip('¡').strip('!').strip('¿').strip('?')
        if len(substring.strip()) > 2:
            personajes.append(substring)

personajes_filtrados = []
for personaje in personajes:
    personaje_model = nlp(personaje)
    is_verb = False
    for token in personaje_model:
        if token.pos_ == 'VERB':
            is_verb = True
    if not is_verb:
        personajes_filtrados.append(personaje)
pers_frec = Counter(personajes_filtrados)
        
```

Volviendo al tema de los datos infiltrados o falsos positivos, los mismos se dan debido a que el modelo que los reconoce como personajes o no está entrenado con un cierto dateset, este dataset no es perfecto y tiene una determinada precisón. Si observamos los falsos positivos podemos observar que comienzan con mayúscula, muchas veces están seguidos de verbos, están al comienzo de la oración o como sujeto, todas características compartidas con los personajes. Un ejemplo sería la siguiente oración: "Jamás ha salido de la casa", como se puede notar un algoritmo automático de inteligencia artificial podría confundir a Jamás con una persona que acaba de salir de la casa. Todo esto hace que el modelo pueda cometer errores y clasificar mal las entidades. Finalmente la lista de personajes queda como sigue:

```python
pers_frec
```

Finalmente se planteó una hipótesis y se quiso comprobarla en base a los datos que se pueden recolectar con spacy. Nuestra hipótesis está relacionada con la relación que hay entre los sustantivos y los verbos en un texto, en este caso puntual "La Isla del Tesoro". Basándonos en las siguientes premisas: un sustantivo en general cumple la función de sujeto, de objeto directo u objeto indirecto; en general cuando tenemos un verbo tendremos un sustantivo realizando la acción; muchas veces la acción recaerá sobre un objeto directo o indirecto; tomando como una aproximación que todo verbo tendrá sujeto y el 50% de los mismos tendrá objeto directo u indirecto planteamos: "En general en un texto, tendremos aproximadamente un 50% más de sustantivos que de verbos". Ahora se procede a contar sustantivos y verbos de nuestro texto para aceptar o rechazar la hipótesis:

```python
verbos = []
sustantivos = []
for token in texto_procesado:
    if token.pos_ == 'VERB':
        verbos.append(token.text)
    if token.pos_ == 'NOUN' and token.is_stop != True and len(token.text) > 2:
        sustantivos.append(token.text)
relacion = len(sustantivos)/len(verbos) - 1
print("En el texto encontramos un " + str(relacion*100) + "% mas de sustantivos que de verbos.")
```

Podemos observar que tenemos un 40% más de sustantivos que de verbos. Si bien no coincide exactamente con el 50% planteado en la hipótesis, lo cual nos lleva a rechazarla, luego de este análisis podemos decir que la relación sustantivo/verbo coincide con la que podemos suponer observando la estructura sintáctica que en general tiene una oración.

Comenzamos con la representación en Python del grafo de la imagen anterior:

#### Grafo como Lista de Aristas

```python
edgeList = [ [0, 2], [1, 3], [2, 3], [2, 4], [3, 5], [4, 5] ];
edgeList
```

#### Grafo como Matriz de Adyacencia

En la siguiente matriz la dirección con la cual se unen los nodos es tomando del número de fila al número de columna:

```python
max_node = 0
for element in edgeList:
    for subelem in element:
        if subelem > max_node:
            max_node = subelem

AdjMtx = np.zeros((max_node + 1, max_node + 1))
AdjList = []
for i in range(max_node + 1):
    AdjList.append([])
for element in edgeList:
    AdjMtx[element[0]][element[1]] = 1
    AdjList[element[0]].append(element[1])
print(AdjMtx)
```
