Graficamos las curvas correspondientes a cada percentil para niños y niñas del __DATASET_1__:

```python
tabla_1.Sex.values
```

Como podemos ver, la __tabla_1__ se encuentra dividida en dos ya que volvemos a tener el nombre de la columna __Sex__. Por lo tanto debemos tener cuidado a la hora de graficar y remover dicha fila problemática.

```python
indice = pd.Index(list(tabla_1.Sex))
indice = indice.get_loc("Sex")
droped = tabla_1[:indice]
sexo_percentil = droped.loc[:indice,['P3','P5','P10','P25','P50','P75','P90','P95','P97']]
sexo_percentil.astype(float).plot();
print('Dataset=1, Sexo=Masculino')
```

```python
indice = pd.Index(list(tabla_1.Sex))
indice = indice.get_loc("Sex")
droped = tabla_1[indice+1:]
sexo_percentil = droped.loc[indice+1:,['P3','P5','P10','P25','P50','P75','P90','P95','P97']]
sexo_percentil.astype(float).plot();
print('Dataset=1, Sexo=Femenino')
```
