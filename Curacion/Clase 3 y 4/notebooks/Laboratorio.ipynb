## 1.3. Despersonalizar datos y guardarlos en un nuevo archivo

Importamos __md5__ para poder despersonalizar nuestros datos.

```python
from hashlib import md5
```

```python
permisos_SF_hash = permisos_SF.copy(deep=True)
permisos_SF_hash
```

El dataset con el cual estamos trabajando, no posee nombres de personas, ni números de documentos, etc. Por lo cual, podríamos despersonalizar el __nombre de la calle__, para que de ésta manera no se pueda indentificar la dirección de la obra.

```python
def hashit(val):
    if isinstance(val, float): 
        return str(val)
    return md5(val.encode('utf-8')).hexdigest()
```

```python
permisos_SF_hash['Street Name'] = permisos_SF_hash['Street Name'].apply(hashit)
permisos_SF_hash
```

## 2.3. Codificar variables

Las variables categóricas deben ser etiquetadas como variables numéricas, no como cadenas.

```python
from sklearn import preprocessing
```

```python
permisos_SF.head(5)
```

Debido a la cantidad de columnas con que cuenta nuestro Dataset se cambiarán solo algunas columnas categóricas a modo de ejemplo, puede observarse al final como estas columnas han pasado a ser numéricas:

```python
column1='Block'
le1 = preprocessing.LabelEncoder()
le1.fit(permisos_SF[column1])
print(dict(zip(le1.classes_,le1.transform(le1.classes_))))
permisos_SF[column1] = le1.transform(permisos_SF[column1]) 

column2='Lot'
le2 = preprocessing.LabelEncoder()
le2.fit(permisos_SF[column2])
print(dict(zip(le2.classes_,le2.transform(le2.classes_))))
permisos_SF[column2] = le2.transform(permisos_SF[column2]) 

column3='Street_Name'
le3 = preprocessing.LabelEncoder()
le3.fit(permisos_SF[column3])
print(dict(zip(le3.classes_,le3.transform(le3.classes_))))
permisos_SF[column3] = le3.transform(permisos_SF[column3]) 

column4='Street_Suffix'
le4 = preprocessing.LabelEncoder()
le4.fit(permisos_SF[column4])
print(dict(zip(le4.classes_,le4.transform(le4.classes_))))
permisos_SF[column4] = le4.transform(permisos_SF[column4]) 

permisos_SF
```

En caso de un análisis para un caso real que se necesitan limpiar los datos este último proceso deberá realizarse para todas las columnas categóricas. Ahora guardaremos el último dateset procesado que hemos obtenido ya que revertiremos el cambio de las variables categóricas para continuar el análisis (en caso de una aplicación real no deberíamos invertirlo ya que necesitamos a las columnas categóricas como números):

```python
permisos_SF.to_csv("../input/Permisos_SF/Building_Permits_cat2num.csv")
```

```python
permisos_SF[column1] = le1.inverse_transform(permisos_SF[column1])
permisos_SF[column2] = le2.inverse_transform(permisos_SF[column2])
permisos_SF[column3] = le3.inverse_transform(permisos_SF[column3])
permisos_SF[column4] = le4.inverse_transform(permisos_SF[column4])

permisos_SF
```
