Se trabajará con un dataset que contiene información de un gran número de empresas de estados unidos con características para cada una de ellas. El objetivo es reducir el número de empresas con las que se trabaja a las relacionadas con el área software. Luego se generará de manera random una columna nueva que diga el puesto de trabajo que la empresa en cuestión está buscando (simulando que las empresas han sido extraídas de un motor de búsqueda determinado). Se provocará que la misma empresa busque por varios puestos de trabajo y a través de algoritmos de clustering se buscará agrupar por empresas suponiendo que el número de empresas es desconocido. Es decir el objetivo es hacer clustering para encontrar todos los post hechos por una misma empresa.

## Data Analysis & Depuración

Primero importamos los datos

```python
import pandas as pd
import numpy as np
import random

random.seed(0)
data = pd.read_csv('Dataset/us_companies.csv')
df = pd.DataFrame(data)
```

```python
pd.set_option('display.max_rows', df.shape[0])
pd.set_option('display.max_columns', df.shape[1])
df
```

```python
df.drop('social_impact', axis=1, inplace=True)
df.drop('source_count', axis=1, inplace=True)
df.drop('data_types', axis=1, inplace=True)
df.drop('example_uses', axis=1, inplace=True)
df.drop('data_impacts', axis=1, inplace=True)
df.drop('financial_info', axis=1, inplace=True)
df.drop('last_updated', axis=1, inplace=True)
```

```python
df
```

Ahora se extraerán columnas cuya información no será de utilidad para los análisis:
- country: es constante para todas las empresas, no aporta información.
- state y city: la información de location ya está aportada por el zipcode. Este dato sobra.
- company name id, company name y url aportan información similar. Para nuestros modelos será de utilidad tomar solo una de estas tres, se tomará company name id que es el más normalizado.

```python
df.drop('country', axis=1, inplace=True)
df.drop('state', axis=1, inplace=True)
df.drop('city', axis=1, inplace=True)
df.drop('company_name', axis=1, inplace=True)
df.drop('url', axis=1, inplace=True)
```

```python
df
```

Estás son finalmente las categorías que interesan para trabajar. Dependiendo la aplicación que se haga se utilizarán solo algunas de ellas. Ahora se hará un poco de curación de los datos. Comenzaremos extrayendo los NaN de las diferentes categorías

```python
def replace_nan_revenue_source(val):
    if str(val) == 'nan':
        return 'Not reported by company'
    else:
        return val

def replace_nan_company_type(val):
    if str(val) == 'nan':
        return 'Private'
    else:
        return val

def replace_nan_company_category(val):
    if str(val) == 'nan':
        return 'Unknown'
    else:
        return val
    
def replace_nan_zip(val):
    if str(val) == 'nan':
        return '0.0'
    else:
        return val
    
def replace_nan_year(val):
    if str(val) == 'nan':
        return '2008.5' # Unknown Value
    else:
        return val
    
def replace_nan_employees(val):
    if val == 0.0:
        return round(df['full_time_employees'].mean(),2)
    else:
        return round(val, 2)
    
def recompute_employees(val):
    if str(val) == 'nan':
        return 0.0
    else:
        val_list = val.split('-')
        if len(val_list)==1:
            return round(float(val_list[0].split('+')[0].replace(',','')), 2)
        else:
            float_list = []
            for element in val_list:
                float_list.append(float(element.replace(',','')))
            np_list = np.array(float_list)
            return round(np_list.mean(), 2)
```

```python
df['revenue_ source'] = df['revenue_source'].apply(replace_nan_revenue_source)
```

```python
df['business_model'] = df['business_model'].apply(replace_nan_revenue_source)
```

```python
df['company_type'] = df['company_type'].apply(replace_nan_company_type)
df['company_category'] = df['company_category'].apply(replace_nan_company_category)
```

```python
df['zip_code'] = df['zip_code'].apply(replace_nan_zip)
df['year_founded'] = df['year_founded'].apply(replace_nan_year)
```

```python
df['full_time_employees'] = df['full_time_employees'].apply(recompute_employees)
```

```python
df['full_time_employees'] = df['full_time_employees'].apply(replace_nan_employees)
```

Ahora el dataset está listo para ser utilizado con un determinado modelo. Pero no se utilizará el dataset completo, se ha curado todo el dataset con el objetivo de poder reutilizarlo ya curado en un futuro. Pero solo se utilizarán 10 empresas, de variados rubros, para simular la búsqueda de los puestos de trabajo. Se seleccionarán las siguientes 10 empresas:
- accuweather
- allianz

- citigroup
- fujitsu
- garmin

- zurich-insurance-risk-room

Se quiere tener una base de 60 puestos de trabajo por lo que se generaran 5 puestos para cada empresa.

```python
companies_list = ['accuweather', 'allianz', 'amazon-web-services', 'citigroup', 'fujitsu', 
                  'garmin', 'github', 'ibm', 'metlife', 'uber', 'yahoo', 
                  'zurich-insurance-risk-room']
jobs = ['Python Software Engineer', 'Java Developer', 'Cloud Engineer', 'Data Scientist']
jobs_column = []

for idx, company in enumerate(companies_list):
    if idx == 0:
        df_companies = df[df['company_name_id']==company]
    else:
        df_companies = df_companies.append(df[df['company_name_id']==company], ignore_index=True)
    random_job = random.randint(0,len(jobs)-1)
    jobs_column.append(jobs[random_job])
    for job in jobs:
        df_companies = df_companies.append(df[df['company_name_id']==company], ignore_index=True)
        jobs_column.append(job)
df_companies
```

```python
df_companies['job_name'] = jobs_column
```

```python
df_companies['job_id'] = [idx for idx,value in enumerate(jobs_column)]
```

```python
df_companies
```

La última columna de id se agregó para diferenciar entre todos los trabajos, en caso que no se utilice será eliminada. Finalmente se obtuvo el set de datos deseado: 60 puestos de trabajo buscados por 10 empresas distintas. Como se puede observar uno de los jobs generados fue generado de manera random, esto es con la intención de que un job se repita por cada empresa. A este job repetido se le borrará la empresa de origen y luego cuando se corra el clustering se evaluará si lo agrupa en el cluster correspondiente.

```python
jobs_to_clean = []
for company in companies_list:
    tmp_jobs_list = []
    for index, row in df_companies[df_companies['company_name_id']==company].iterrows():
        if row['job_name'] in tmp_jobs_list:
            jobs_to_clean.append(row['job_id'])
        tmp_jobs_list.append(row['job_name'])
```

```python
for job_id in jobs_to_clean:
    df_companies.loc[job_id,'company_name_id'] = ' '
```
