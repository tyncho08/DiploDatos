Se diseñará ahora una métrica de evaluación basada en testigos para poder tener una idea de que tan bien está andando nuestro algoritmo. Además servirá para estimar el número óptimo de clusters. 

```python
# Computo de la precision del algoritmo en base a un testigo
def compute_cluster_score(cluster, testigo):
    clases_dict = {}
    for post in cluster:
        label = post[testigo][post.index.values[0]]
        if not str(label) in clases_dict.keys():
            clases_dict[str(label)] = []
            clases_dict[str(label)].append(label)
        else:
            clases_dict[str(label)].append(label)
    bigger_class = 0
    for element in clases_dict:
        if len(clases_dict[element]) > bigger_class:
            bigger_class = len(clases_dict[element])
    
    return (bigger_class*1.0)/len(cluster)     
    
        
```

Se genera un diccionario con los diferentes clusters calculados. Cada cluster posee adentro una lista de Dataframes, y cada Dataframe corresponde a una línea del DataFrame original. Este diccionario nos sirve para poder tener toda la información separada en los correspondientes clusters.

```python
clusters_dict = {}
for idx, label in enumerate(labels):
    if not str(label) in clusters_dict.keys():
        clusters_dict[str(label)] = []
        clusters_dict[str(label)].append(df_original.loc[[idx]])
    else:
        clusters_dict[str(label)].append(df_original.loc[[idx]])
```

Ahora se procede a realizar la evaluación por testigos. Debido a que no se tiene ninguna feature que sea totalmente característica de un solo cluster y por lo tanto se pueden repetir las mismas de un cluster a otro, se elegirán dos features como testigos ya que hay muchas menos probabilidades de que se repitan las dos juntas. Se eligieron como features testigos para evaluar si los clusters están bien agrupados las features: year_founded y zip_code.
La función compute_cluster_score() toma el testigo y se fija el value mayoritario para esa categoría, y luego retorna la relación entre la cantidad de estos value sobre el total de elementos del cluster. Es decir si el cluster está correctamente agrupado el score debe dar 1. El score total de cada cluster será el menor de los dos calculados.

```python
for cluster in clusters_dict:
    score_1 = compute_cluster_score(clusters_dict[cluster], 'year_founded')
    score_2 = compute_cluster_score(clusters_dict[cluster], 'zip_code')
    if score_1 < score_2:
        score = score_1
    else:
        score = score_2
    print("Score para Cluster "+ str(cluster)+": " + 
          str(score))
```

Cómo se puede observar el algoritmo ha agrupado a los puestos de trabajo tal cual se esperaba. Y las métricas de evaluación hasta este punto parecen funcionar correctamente. Este análisis fue hecho conociendo el número de clusters en que se debería dividir el conjunto. A continuación se utilizará la métrica de evaluación para encontrar el número de clusters óptimo, suponiendo que es desconocido:

```python
num_clusters = 1
mean_score = 0

while(mean_score < 0.99):
    num_clusters += 1
    print('Prueba con numero de clusters: ' + str(num_clusters))
    km = KMeans(n_clusters=num_clusters)
    km.fit(df)

    clusters = km.labels_.tolist()

    # Recuento del número de elementos en cada cluster
    for i in range(num_clusters):
        print ('    El cluster %i tiene %i elementos' % (i, clusters.count(i)))

    # Get cluster assignment labels
    labels = km.labels_

    clusters_dict = {}
    for idx, label in enumerate(labels):
        if not str(label) in clusters_dict.keys():
            clusters_dict[str(label)] = []
            clusters_dict[str(label)].append(df_original.loc[[idx]])
        else:
            clusters_dict[str(label)].append(df_original.loc[[idx]])

    total_score = 0
    for cluster in clusters_dict:
        score_1 = compute_cluster_score(clusters_dict[cluster], 'year_founded')
        score_2 = compute_cluster_score(clusters_dict[cluster], 'zip_code')
        if score_1 < score_2:
            score = score_1
        else:
            score = score_2
        print("    Score para Cluster "+ str(cluster)+": " + 
              str(score))
        total_score = total_score + score
    mean_score = (total_score * 1.0) / num_clusters
    print('    Score Promedio: ' + str(mean_score))
```

Se comienza variando el número de veces en que se inicializan las semillas (por default es 10) y el número de iteraciones (default 300). Si utilizamos K = 12 no se podrá observar ningún cambió, por lo que se utiliza K = 10 para poder hacer un análisis comparando con la corrida anterior.

```python
param_grid = {
    'n_clusters': [10],
    'n_init': [20, 30, 40],
    'max_iter': [500, 1000],
    }

params_list = list(ParameterGrid(param_grid))

for params in params_list:
    print('Parametros: ' + str(params))
    num_clusters = params['n_clusters']
    km = KMeans()
    km.set_params(**params)
    km.fit(df)

    clusters = km.labels_.tolist()

    # Recuento del número de elementos en cada cluster
    for i in range(num_clusters):
        print ('    El cluster %i tiene %i elementos' % (i, clusters.count(i)))

    # Get cluster assignment labels
    labels = km.labels_

    clusters_dict = {}
    for idx, label in enumerate(labels):
        if not str(label) in clusters_dict.keys():
            clusters_dict[str(label)] = []
            clusters_dict[str(label)].append(df_original.loc[[idx]])
        else:
            clusters_dict[str(label)].append(df_original.loc[[idx]])

    total_score = 0
    for cluster in clusters_dict:
        score_1 = compute_cluster_score(clusters_dict[cluster], 'year_founded')
        score_2 = compute_cluster_score(clusters_dict[cluster], 'zip_code')
        if score_1 < score_2:
            score = score_1
        else:
            score = score_2
        print("    Score para Cluster "+ str(cluster)+": " + 
              str(score))
        total_score = total_score + score
    mean_score = (total_score * 1.0) / num_clusters
    print('    Score Promedio: ' + str(mean_score))
```

En un caso real por supuesto, con datos más ruidosos, y con un volumen de datos mayor, es muy dificil que este score de 1 y habría que fijar probablemente el umbral para encontrar el número óptimo de clusters (actualmente 0.99) en un número menor. Como se puede observar para el número de clusters igual a 12, los datos jobs se distribuyen en forma pareja dentro de cada cluster (5 por cluster), esto tiene sentido ya que se habían generado 5 job posts por compañía. Además puede observarse como se reasignan estos jobs a otros clusters, por ejemplo con K = 10, se tendrán "dos compañías menos", es decir el algoritmo va a distribuir los datos de dos compañías en otros clusters en base a determinadas similitudes.
Cómo era de esperar, al variar los parámetros no se ha obtenido diferencia en el score total para cada conjunto de parámetros. Si se puede ver que el score individual de cada cluster cambia. Esto es porque estamos cambiando el número de veces en que se inicializan las semillas, y además no le fijamos el random_seed, lo cual genera que para cada corrida el algoritmo agrupe en base a diferentes similitudes. Además de agrupar por diferentes similitudes, los clusters que se forman no tendrán en cada corrida los mismos elementos. Para evitar esto deberíamos fijar el random_seed para que corra siempre de la misma manera. Notese que para un mismo K el número de elementos reasignados es siempre el mismo. Cómo se ha podido observar al disminuir a 10 clusters, los elementos son reasignados siempre sobre los clusters 1, 3, 5 y 9. ¿Por qué se da esto? el algoritmo se ve forzado a encontrar similitudes entre jobs de diferentes compañías ya que el número en que agrupa los datos se ha reducido.

## Elementos mas cercanos al centro ##
Lo que se utiliza para el calculo de los elementos del cluster más cercanos al centro es la distancia coseno. Para esto se calcula la distancia entre cada elemento de cada cluster al centroide correspondiente, y luego se imprime el índice de los n más cercanos. Ahora el diccionario de los clusters en lugar de armarse con el dataframe original se armará con los valores a partir de los cuales se hizo la clusterización (todos categóricos) para poder calcular la distancia de los elementos de este diccionario. Luego con el índice siempre se puede volver al dataset original.

```python
def row_to_dict(row):
    tmp_list = []
    tmp_list.append(row.company_name_id[row.company_name_id.index[0]])
    tmp_list.append(row.year_founded[row.year_founded.index[0]])
    tmp_list.append(row.zip_code[row.zip_code.index[0]])
    tmp_list.append(row.full_time_employees[row.full_time_employees.index[0]])
    tmp_list.append(row.company_category[row.company_category.index[0]])
    tmp_list.append(row.job_name[row.job_name.index[0]])
    return np.array(tmp_list)

def compute_distance(element1, element2):
    return 1 - spatial.distance.cosine(element1, element2)

    
```
